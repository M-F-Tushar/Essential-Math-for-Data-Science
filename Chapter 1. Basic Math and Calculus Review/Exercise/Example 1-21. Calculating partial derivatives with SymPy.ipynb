Calculate the partial derivatives
for x and y, respectively, with SymPy.

```python
from sympy import *
from sympy.plotting import plot3d
```

```python
# Declare x and y to sympy
x, y = symbols('x y')
```

```python
# Now just use python syntax to declare function
f = 2*x**3 + 3*x**3
```

```python
# Calculate the partial derivatives for x and y
dx_f = diff(f, x)
dx_y = diff(f, y)
```

```python
print(dx_f) # print the partial derivative of f with respect to x
print(dx_y) # print the partial derivative of f with respect to y
```

```python
# plot the function
plot3d(f)
```
